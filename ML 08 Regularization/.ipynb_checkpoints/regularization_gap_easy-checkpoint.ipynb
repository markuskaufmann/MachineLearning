### Regularization for Simple Linear Regresion

Although linear regression is a linear machine learning method, you can have nonlinear dependencies if you transform some of the independent variables by a nonlinear function. By doing this, you can improve the fit of your method. Let us demonstrate this on a house price dataset. Note that this dataset is not identical with one you used in the linear regression exercise, since the those datset is too small and would cause unreliable evaluation results. 

```python
dforig=pd.read_csv("kc_house_data.csv")
```

```python
dforig
```

We would like to have a simple linear regression problem with only one independent variable. Thus, we only keep price and sqft_living.

```python
df=dforig
df=df[["price","sqft_living"]]
#del df["Bedrooms"]
```

To account for the bias term, we add a column containing only ones.

```python
num_rows=df.shape[0]
ones=[1]*num_rows
df["bias"]=ones
```

```python
df
```

We split the data into training and test set

```python
train, test = train_test_split(df, test_size=0.5)
```

We define a linear regression function based on the normal equation 

```python
def linear_regression(X,y):
    Xt=np.transpose(X)
    XtX=np.matmul(Xt,X)
    XtXm1=np.linalg.pinv(XtX)
    Xty=np.matmul(Xt,y)
    thetas=np.matmul(XtXm1,Xty)
    return thetas
```

Let us normalize the variables.

```python
max_value = train["sqft_living"].max()
min_value = train["sqft_living"].min()
train["sqft_living"] = (train["sqft_living"] - min_value) / (max_value - min_value)
X=train[["bias","sqft_living"]]

test["sqft_living"]=(test["sqft_living"]-min_value) / (max_value - min_value)

max_value=train["price"].max()
min_value=train["price"].min()
train["price"] = (train["price"] - min_value) / (max_value - min_value)
test["price"] = (test["price"] - min_value) / (max_value - min_value)

```

```python
train
```

```python
y=train[["price"]]
```

```python
y
```

We apply linear regression to obtain the optimal parameters $\theta$.

```python
thetas=linear_regression(X,y)
```

```python
thetas
```

Using X and the estimated thetas, we can predict the house prices by a straight line

```python
ypred=np.matmul(X,thetas)
```

```python
ypred
```

Let us plot house prices and predicted house prices

```python
plt.plot(train["sqft_living"],y,'bo',markersize=1)
xs = np.arange(0,1.1,0.01)
lenxs=xs.shape[0]
ones=[1]*lenxs
Xnt=np.stack((ones,xs))
Xn=np.transpose(Xnt)
ypred=np.matmul(Xn,thetas)
plt.plot(xs,ypred)
```

We define a convenience method that helps us to stack matrix columns containing different powers of x next to each other.

```python
def stack_xn(ones,xs,n):
    li=[]
    li.append(ones)
    xss=xs
    for i in range(0,n):
        li.append(xss)
        xss=xss*xs
    return np.transpose(np.stack(li))

```

Now let us check how the obtained parameter perform on the test set.

```python
num_rows=test.shape[0]
ones_test=[1]*num_rows
X_testt=np.stack((ones_test,test["sqft_living"]))
X_test=np.transpose(X_testt)

y_pred_test=np.matmul(X_test,thetas)
y_test=test["price"]
y_pred_test=y_pred_test.reshape((-1))
r2= r2_score(y_test, y_pred_test)
print ("R2: ",r2)
```

Now we aim to improve the fit by adding $x^2$ as additional independent variable.

```python
dfx2=train
dfx2["sqft_living^2"]=dfx2["sqft_living"]*dfx2["sqft_living"]

X=dfx2[["bias","sqft_living","sqft_living^2"]]
thetas=linear_regression(X,y)
print ("thetas: ",thetas)
Xnt=np.stack((ones,xs,xs*xs))
Xn=np.transpose(Xnt)
print ("Xn: ",Xn)
ypred=np.matmul(Xn,thetas)
plt.figure(figsize=(15,10))
plt.plot(dfx2["sqft_living"],y,'bo',markersize=2)
plt.plot(xs,ypred)
```

Again, we determine the R2 score on the test set. 

```python
num_rows=test.shape[0]
onestest=[1]*num_rows
xstest=test["sqft_living"]
X_test=stack_xn(onestest,xstest,2)

y_pred_test=np.matmul(X_test,thetas)
y_test=test["price"]
y_pred_test=y_pred_test.reshape((-1))
r2= r2_score(y_test, y_pred_test)
print ("R2: ",r2)
```

If we add higher order polynomials, the graph begins to overfit. We are adding $x^3$, $x^4$, and $x^5$ as 3 additional independent variables.

```python
dfx2["sqft_living^3"]=dfx2["sqft_living^2"]*dfx2["sqft_living"]
dfx2["sqft_living^4"]=dfx2["sqft_living^3"]*dfx2["sqft_living"]
dfx2["sqft_living^5"]=dfx2["sqft_living^4"]*dfx2["sqft_living"]
dfx2["sqft_living^6"]=dfx2["sqft_living^5"]*dfx2["sqft_living"]
X=dfx2[["bias","sqft_living","sqft_living^2","sqft_living^3","sqft_living^4","sqft_living^5"]]
thetas=linear_regression(X,y)
print ("thetas: ",thetas)
Xn=stack_xn(ones,xs,5)
ypred=np.matmul(Xn,thetas)
plt.figure(figsize=(15,10))
plt.xlim(0,1.1)
plt.ylim(0, 1.1)
plt.plot(dfx2["sqft_living"],y,'bo',markersize=2)
plt.plot(xs,ypred)


```

Now again, determine the R2 for the new parameter set $\theta$ on the test set.
You see, that often, the $R^2$ is already worse than for the previous parameter vector.

```python
num_rows=test.shape[0]
ones_test=[1]*num_rows
xstest=test["sqft_living"]
X_test=stack_xn(ones_test,xstest,5)

y_pred_test=np.matmul(X_test,thetas)
y_test=test["price"]
y_pred_test=y_pred_test.reshape((-1))
r2= r2_score(y_test, y_pred_test)
print ("R2: ",r2)
```

The effect of overfitting can be reduced by regularization. <font color="red">Complete the implementation of the regularized version of linear regression:

$$\boldsymbol\theta := \left(\mathbf{X}^T\mathbf{X}+\lambda \begin{bmatrix}
    0      & 0 &\cdots&0 \\
    0      &   1 & \cdots   &  0\\
    \vdots & \vdots & \ddots & \vdots\\
    0      & 0  & \cdots & 1
  \end{bmatrix} \right)^{-1}\mathbf{X}^T \mathbf{y}
$$

</font> 

```python
def linear_regression_reg(X,y,lam):
    Xt=np.transpose(X)
    XtX=np.matmul(Xt,X)
    num_rows=XtX.shape[0]
    
    # Add lambda to the diagonal elements of X'X except to the first (see formula above)
    for i in range(1,num_rows):
        XtX[i,i] = XtX[i,i] + lam   #Your code
        
    XtXm1=np.linalg.pinv(XtX)
    Xty=np.matmul(Xt,y)
    thetas=np.matmul(XtXm1,Xty)
    return thetas
```

We now plot the graph using the regularized parameter vectors. As you can see, the effect of overfitting is strongly reduced.

```python
thetas=linear_regression_reg(X,y,4)
```

```python
thetas
```

```python
ypred=np.matmul(Xn,thetas)

plt.figure(figsize=(15,10))
plt.xlim(0,1)
plt.ylim(0, 1)
plt.plot(dfx2["sqft_living"],y,'bo')
plt.plot(xs,ypred)
```

Let us determine the $R^2$ using regularization.

```python
num_rows=test.shape[0]
ones_test=[1]*num_rows
xstest=test["sqft_living"]
X_test=stack_xn(ones_test,xstest,10)


y_pred_test=np.matmul(X_test,thetas)
y_test=test["price"]
y_pred_test=y_pred_test.reshape((-1))
r2= r2_score(y_test, y_pred_test)
print ("R2: ",r2)
```

You should see that the $R^2$ is much higher than for the unregularized parameter set.

### Regularization with logistic regression 

Now let us apply regularization to a larger dataset using sklearn, for which we reuse the ezcema dataset from the last exercise.

```python
df=pd.read_csv("skin_disease.csv")
di = {1: 0, 255: 2}
df=df.replace({"class": di})
di = {2: 1}
df=df.replace({"class": di})
del df["x"]
del df["y"]
del df["expertsEczemaVotePatch"]
del df["expertsEczemaAllVotePatch"]
del df["expertsEczemaVoteCenterPixel"]
```

We again split the data into training and test set.

```python
train, test = train_test_split(df, test_size=0.2)
X_train = train.drop('class', axis=1)
X_test = test.drop('class', axis=1)

y_train = train["class"]
y_test = test["class"]
```

<font color="red">You should systematically try out all L2 regularization parameters in the interval [0,2] with step size 0.1, a technique also known as grid search. Hint: Use the class $\texttt{sklearn.linear_model.LogisticRegression}$ and specify the regularization type with the parameter penalty.</font>

```python

for i in range(1,21):
    c=i*0.1
    logReg=sklearn.linear_model.LogisticRegression(penalty='l2',C=c)
    logReg.fit(X_train,y_train)
    y_pred_test=logReg.predict(X_test)
    accuracy=accuracy_score(y_test, y_pred_test)
    fscore=f1_score(y_test,y_pred_test)
    print ("C: ",c," accuracy: ",accuracy, "f1 score: ",fscore)
    
```

One final remark, in this notebook we used for simplicity skin pixels from the same image for training and testing.
In real world scenarios, the test and training data set should originate from different images, 
since skin pixels from the same image are highly correlated and your evaluation measure values would therefore 
probably be too high.
